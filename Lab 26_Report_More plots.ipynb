‚úÖBar Chart grouped for comparison demographic 2023 vs 2024 Survey

Dataframe was created using the % results from 2023 and 2024 surveys

[2023 Survey](https://survey.stackoverflow.co/2023/)

[2024 Survey](https://survey.stackoverflow.co/2024/)

```python
89184-65437
```

```python
# Data
years = ["2023", "2024"]
participants = [89184, 65437]

# Bar Chart
plt.bar(years, participants, color=["#fbaa6e", "#fa8631"])

# Annotation pour la diff√©rence
plt.text(1.1, (participants[0] + participants[1]) / 2, "-23 747", fontsize=14, color="#3b3a39", ha="right")
plt.text(1.1, (participants[0] + participants[1]) / 2.2, "-26.6%", fontsize=14, color="#fa2702", ha="right")
# Formatage
plt.ylabel("Number of Respondents")
plt.title("Comparison of Survey Respondents 2023 vs. 2024", fontsize=16)
plt.tight_layout()
plt.show()
```

```python
demographic = {
    "Criteria": [
        "USA Participation", "Bachelor Degree", "Master Degree",
        "25-34 years old", "35-44 years old", "18-24 years old", "Employed full-time", "Freelancer", "Student full-time"
    ],            
    "2023": [21.21, 41.16,23.03,37.28,23.02,20.11,69.28,15.91,13.3],  
    "2024": [18.9,41,25.6,36.5,22.9,21.4,69,16.4,13.2]}

# Create df
data = pd.DataFrame(demographic)

print(data)
```

```python
# Cr√©ation du graphique
criteria = data["Criteria"]
values_2023 = data["2023"]
values_2024 = data["2024"]
fig=plt.subplots(figsize=(12, 6))
x = range(len(criteria))
width = 0.4  

plt.bar(x, values_2023, width=width, label='2023',color='#fbaa6e', edgecolor='#616a6b')
plt.bar([i + width for i in x], values_2024, width=width, label='2024',color='#fa8631', edgecolor='#616a6b')

# Configuration des √©tiquettes des x
# Ajustement des √©tiquettes pour l'axe des x
labels = []
for i in range(len(criteria)):
    if i % 2 == 0:  # Crit√®res pairs
        labels.append("\n" + criteria[i])  # Ajoute un saut de ligne pour abaisser l'√©tiquette
    else:  # Crit√®res impairs
        labels.append(criteria[i])  # Garde les √©tiquettes au niveau normal

plt.xticks([i + width / 2 for i in x], labels, fontsize=12, ha="center")

plt.ylabel("Proportion (%)",fontsize=14)
plt.title("Survey Demographics 2023 vs 2024",fontsize=18)
plt.legend(fontsize=14)
```
